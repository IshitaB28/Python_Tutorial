```python
df1=pd.DataFrame({'key':['k0', 'k1', 'k2', 'k3'],
             'x': [1,2,3,4],
             'y': [4,5,6,7]})
df1
```

```python
df2=pd.DataFrame({'key':['k0', 'k6', 'k2', 'k8'],
             'a': [1,2,3,4],
             'b': [4,5,6,7]})
df2
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, how='left')
# how determines inner, left, right or outer join operations
#by default inner
```

```python
pd.merge(df1, df2, how='right', left_on='x', right_on='b')
```

```python
df1.set_index('key', inplace=True)
df1
```

```python
df2.set_index('key', inplace=True)
df2
```

```python
# above operations were merge operations. Now we see the join operations
#join operations is basically done on the indexes
```

```python
df1.join(df2)  # left join by default
```

```python
df1.join(df2, how='right')
```

```python
df1.join(df2, how='outer')
```
